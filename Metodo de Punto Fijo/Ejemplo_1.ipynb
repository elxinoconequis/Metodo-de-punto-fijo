```python
import numpy as np # sirve para hacer arreglos y operaciones entre arreglos
import matplotlib.pyplot as plt # nos sirve para grafica
import scipy.stats as stats #sirve para estadistica
import math # sirve para hacer funciones matematicas, como seno, coseno, potencias, etc.

# Ejemplo 2.1
x = np.linspace(-4,4,50)
y = np.sin(x)
z = x
t = np.zeros(len(x)) # len(): es una función que saca el tamaño de un arreglo viende de 'length'
plt.rcParams['figure.dpi'] = 200 # cambiar tamaño del gráfico
plt.plot(x, y)
plt.plot(x,z)
plt.plot(x,t)
plt.legend(["sin(x)","x","0"])
x_min = -4
x_max = 4
y_min = -2
y_max = 2
plt.axis([x_min, x_max, y_min, y_max])
plt.show()
```

```python
iteraciones = 1000
i = np.arange(0,iteraciones+1)
print(i)
x_ = np.zeros(len(i))# tiene once elementos
g_x_ = np.zeros(len(i))
abs_f_x_ = np.zeros(len(i))
x_[0]= 2 # Valor inicial
n = 0 # variable auxiliar

for n in range(len(i)-1): # estamos haciendo n-1 iteraciones, para no romper la indexación del vector
    print("El valor de 'n': ", n)
    g_x_[n] = np.sin(x_[n])
    print("El valor de g(x)= ", g_x_[n])
    # El siguiente paso es sustituir el nuevo valor de 'x[n]'
    x_[n+1] = g_x_[n]
    # Vamos a evaluar f(x)
    abs_f_x_[n] = x_[n]-(np.sin(x_[n]))
    print("El valor de f(x)", abs(abs_f_x_[n]))

print(x_)
```
