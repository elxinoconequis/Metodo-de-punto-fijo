La primera forma alternativa de la ecuación que se propuso es:
    
$
\begin{align}
\large g_{1}(x)=x=\sqrt[2]\frac{{e^{x}}}{cos(x)} \label{Ec.2}
\end{align}
$

```python
def Eq1_forma1(Valor_inicial: float, iteraciones: int) -> pd.DataFrame:
    i = np.arange(0,iteraciones+1)
    #print(i)
    
    # Variables para caso real
    g_x_ = np.zeros(len(i))
    x_ = np.zeros(len(i))
    f_x_ = np.zeros(len(i))
    abs_f_x_ = np.zeros(len(i))
    
    # Variable spara el caso imaginario
    array_size = iteraciones + 1
    g_x_j_ = [complex()] * array_size
    x_j_ = [complex()] * array_size
    f_x_j_ = [complex()] * array_size
    abs_f_x_j_ = [complex()] * array_size
    
    # Valor inicial para ambos casos
    x_[0] = Valor_inicial
    x_j_[0] = Valor_inicial
    
    for n in range(len(i) - 1):
        #print(n)
        a = (mt.e**x[n])
        b = (mt.cos(x[n]))
        try:
            #print("usando 'math'")
            c = a / b
            #print(c)
            g_x_[n] = mt.sqrt(c)
            x_[n + 1] = g_x_[n] 
            f_x_[n] = (x_[n]**2)*(mt.cos(x_[n])) - ((mt.e)**(x_[n]))
            abs_f_x_ = abs(f_x_[n])
        except:
            #print("usando 'cmath'\n    ")
            a = (cmt.e**x[n])
            b = (cmt.cos(x[n]))
            c = a / b
            #print(c)
            g_x_j_[n] = cmt.sqrt(c)
            x_j_[n + 1] = g_x_j_[n]
            f_x_j_[n] = (x_[n]**2)*(cmt.cos(x_[n])) - ((cmt.e)**(x_[n]))
            
            
            
    data_real = {"$x_{i}$": x_, "$g(x_{i})$":g_x_,"$f(x)":f_x_,"$\mid f(x_{i} \mid)$":abs_f_x_}
    data_j = {"$x_{i}$": x_j_, "$g(x_{i})$":g_x_j_,"$f(x)":f_x_j_,"$\mid f(x_{i} \mid)$":abs_f_x_j_}
    Tabla_resultados_reales = pd.DataFrame(data_real)
    Tabla_resultados_complejos = pd.DataFrame(data_j)
    return Tabla_resultados_reales, Tabla_resultados_complejos
```

El problema de este último rpoblema es que inmediatemente entramos con número complejos la tratar de hallar una solución
Por eso voy a proponer una forma alternativa
$
\begin{align}
\large g_{2}(x)=x^{2}cos(x)+x-e^{x} = x \label{Ec. 3}
\end{align}
$


```python
def Eq1_forma2(Valor_inicial: float, iteraciones: int) -> pd.DataFrame:
    i = np.arange(0,iteraciones+1)
    # Variables para caso real
    g_x_ = np.zeros(len(i))
    x_ = np.zeros(len(i))
    f_x_ = np.zeros(len(i))
    abs_f_x_ = np.zeros(len(i))
    # Valor incial
    x_[0] = Valor_inicial
    # Algortimo
    for n in range(len(i)-1):
        g_x_[n] = (x_[n]**2)*(np.cos(x_[n])) + x_[n] - (np.exp(x_[n]))
        x_[n + 1] = g_x_[n]
        f_x_[n] = (x_[n]**2)*(np.cos(x_[n])) - (np.exp(x_[n]))
        abs_f_x_[n] = abs(f_x_[n])
        
    
    data_real = {"$x_{i}$": x_, "$g(x_{i})$":g_x_,"$f(x)$":f_x_,"$\mid f(x_{i} \mid)$":abs_f_x_}
    Tabla_resultados_reales = pd.DataFrame(data_real)
    return Tabla_resultados_reales
```

Forma alternativa 1 de la ecuación 2
$$
g_{1}(x) = tan^{-1}(ln(x))
$$

```python
def Eq2_forma1(Valor_inicial: float, iteraciones: int) -> pd.DataFrame:
    i = np.arange(0,iteraciones+1)
    # Variables para caso real
    g_x_ = np.zeros(len(i))
    x_ = np.zeros(len(i))
    f_x_ = np.zeros(len(i))
    abs_f_x_ = np.zeros(len(i))
    # Valor incial
    x_[0] = Valor_inicial
    # Algortimo
    for n in range(len(i)-1):
        a = np.log(x_[n])
        g_x_[n] = np.arctan(a)
        x_[n + 1] = g_x_[n]
        f_x_[n] = np.tan(x_[n]) - (np.log(x_[n]))
        abs_f_x_[n] = abs(f_x_[n])
        
    
    data_real = {"$x_{i}$": x_, "$g(x_{i})$":g_x_,"$f(x)$":f_x_,"$\mid f(x_{i} \mid)$":abs_f_x_}
    Tabla_resultados_reales = pd.DataFrame(data_real)
    return Tabla_resultados_reales
```
